```python
import requests
import bs4
from datetime import datetime as dt
import re
```

```python
year = 2017
month = 5
URL = 'http://npb.jp/games/' + str(year) + '/' + 'schedule_' + str('{0:02d}'.format(month)) + '_detail.html'
```

```python
res = requests.get(URL)
res.raise_for_status()
print(res.text)
```

```python
soup = bs4.BeautifulSoup(res.content, "html.parser")
```

```python
date = dt(2017,5,12)
```

```python
results = soup.find_all('tr', id='date'+'{0:02d}'.format(date.month)+'{0:02d}'.format(date.day))
```

```python
len(results)
results
```

```python
dict = {}
dict['date'] = date
for result in results:
    if (result.select('.team1')[0].getText() == '広島'):
        print(result.select('.score1'))
        dict['score'] = result.select('.score1')[0].getText()
        dict['opponent_name'] = result.select('.team2')[0].getText()
        dict['opponent_score'] = result.select('.score2')[0].getText()
    if result.select('.team2')[0].getText() == '広島':
        dict['score'] = result.select('.score2')[0].getText()
        dict['opponent_name'] = result.select('.team1')[0].getText()
        dict['opponent_score'] = result.select('.score1')[0].getText()
```
